# Filtering Data based on conditions

```python
od.loc[od['Weight']<50]
```

```python
od.loc[(od['Weight']<50) & (od['Obesity']=='Normal_Weight')] # and condition and be carefull for the indentation
```

```python
od.loc[(od['Weight']<50) | (od['Obesity']=='Normal_Weight')]
```

# CAN ALSO BE USED WITHOUT LOC FOR THE FILTERATION
# CALLED BOOLEAN INDEXING

```python
od[(od['Weight']<50) | (od['Obesity']=='Normal_Weight')]
```
