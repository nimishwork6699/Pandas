```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['AJAY','RAM','RAJU','SUMAN']
})
df2=pd.DataFrame({
    'id':[3,4,5,6],
    'courses':['C++','DB','PANDA','PIP']
})
inner_result= pd.merge(df1,df2,how='inner',on='id') # it is for the comman ones or we can say as intersection
inner_result
```

```python
outer_result=pd.merge(df1,df2,how='outer',on='id') # it will show all the ones or can be said as union
outer_result
```

```python
right_result=pd.merge(df1,df2,how='right',on='id')
right_result
```

```python
left_result=pd.merge(df1,df2,how='left',on='id')
left_result
```

```python
df3=pd.DataFrame({
    'id1':[1,2,3,4],
    'name':['AJAY','RAM','RAJU','SUMAN'] # used for if we have two id from diff csv
})
df4=pd.DataFrame({
    'id':[3,4,5,6],
    'courses':['C++','DB','PANDA','PIP']})

inner_result= pd.merge(df3,df4,how='inner',left_on='id1',right_on='id') # for other left right outer also
inner_result
```

```python
df1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['AJAY','RAM','RAJU','SUMAN']
})
df2=pd.DataFrame({
    'id':[3,4,5,6],
    'courses':['C++','DB','PANDA','PIP']
})
result=pd.concat([df1,df2],axis=0,ignore_index=True) # if we dont do it it will show not in 1 series {ignore_index=True}
result
```
